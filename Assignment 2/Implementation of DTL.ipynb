### Function to calculate Entropy of a specific attribute

```python
def calculate_entropy(labels):
    unique_labels = set(labels)
    entropy = 0
    for label in unique_labels:
        prob = (labels == label).sum() / len(labels)  # Updated line
        entropy -= prob * math.log2(prob) if prob != 0 else 0  # Handle case when prob is 0
    return entropy
```

### Function to calculate Gain for a specific attribute

```python
def calculate_info_gain(data, target, attribute):
    # Calculate the overall entropy
    total_entropy = calculate_entropy(data[target])

    # Calculate the weighted entropy for each value of the attribute
    values = set(data[attribute])
    weighted_entropy = 0
    for value in values:
        subset = data[data[attribute] == value]
        subset_entropy = calculate_entropy(subset[target])
        weight = len(subset) / len(data)
        weighted_entropy += weight * subset_entropy

    # Calculate information gain
    info_gain = total_entropy - weighted_entropy

    return info_gain
```

```python
# Calculate entropy and information gain for each attribute
target_attribute = 'Play Tennis'
attributes = ['Outlook', 'Temperature', 'Humidity', 'Wind']
```

```python
for attribute in attributes:
    entropy = calculate_entropy(df[target_attribute])
    info_gain = calculate_info_gain(df, target_attribute, attribute)
    print(f'{attribute} - Entropy: {entropy:.4f}, Gain: {info_gain:.4f}')
```
